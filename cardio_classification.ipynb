```python
import numpy as np
import pandas as pd
```

```python
fh = pd.read_csv('C:/Users/tsfra/OneDrive/Documents/data622/datasets/fetal_health.csv')
fh['fh_int'] = fh['fetal_health'].astype(int)
```

```python
display(fh[fh.isna()].sum())
```

```python
fh.loc[fh['fh_int'] == 1, 'fetal_health'] = 'Normal'
fh.loc[fh['fh_int'] == 2, 'fetal_health'] = 'Suspect'
fh.loc[fh['fh_int'] == 3, 'fetal_health'] = 'Pathological'

fh['fetal_health'] = pd.Categorical(fh['fetal_health'], 
                                    ordered=True)
```

```python
import plotly.express as px

fig = px.box(fh, 
                y=fh.columns[0], 
                color='fetal_health', 
                category_orders={
                    'fetal_health': ['Normal', 'Suspect', 'Pathological']
                    }
            )
fig.update_layout(title=f"{fh.columns[0]}")
fig.update_yaxes(title=None)
fig.show()

#for i in range(0, len(fh.columns) - 2):
#    fig = px.box(fh, 
#                 y=fh.columns[i], 
#                 color='fetal_health', 
#                 category_orders={
#                     'fetal_health': ['Normal', 'Suspect', 'Pathological']
#                     }
#                )
#    fig.update_layout(title=f"{fh.columns[i]}")
#    fig.update_yaxes(title=None)
#    fig.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = fh.loc[:, [col for col in fh if col != 'fetal_health']].values
y = fh['fh_int'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train, y_train)

gnb_predict = gnb.predict(X_test)
```

```python
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
)

results_df = pd.DataFrame({"Actual": y_test, "Predicted": gnb_predict})
display(results_df[results_df["Actual"] != results_df["Predicted"]].head(10))

gnb_accuracy = accuracy_score(gnb_predict, y_test)
f1 = f1_score(gnb_predict, y_test, average="weighted")

print("Accuracy: ", gnb_accuracy)
print("F1 Score: ", f1)
```

```python
labels = [1, 2, 3]
cm = confusion_matrix(y_test, gnb_predict, labels=labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot()
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score

model = GaussianNB()
model.fit(X, y)

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=0)

scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
print(np.mean(scores))
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

#sc = StandardScaler()
#X_train = sc.fit_transform(X_train)
#X_test = sc.transform(X_test)

lda = LDA(n_components=1)
lda.fit(X_train, y_train)
lda_predict = lda.predict(X_test)
```

```python
df = pd.DataFrame({'Xtrain':X_train[:, 0], 'ytrain':y_train})

fig = px.box(df, 
                y='Xtrain', 
                color='ytrain'
            )
fig.update_yaxes(title=None)
fig.show()
```

```python
results_df = pd.DataFrame({"Actual": y_test, "Predicted": lda_predict})
display(results_df[results_df["Actual"] != results_df["Predicted"]].head(10))

lda_accuracy = accuracy_score(lda_predict, y_test)
f1 = f1_score(lda_predict, y_test, average="weighted")

print("Accuracy: ", lda_accuracy)
print("F1 Score: ", f1)
```

```python
labels = [1, 2, 3]
cm = confusion_matrix(y_test, lda_predict, labels=labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot()
```

```python
model = LDA()
model.fit(X, y)

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
print(np.mean(scores))
```
